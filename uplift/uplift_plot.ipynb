```python
from __future__ import division, unicode_literals, print_function  
import numpy as np
import matplotlib.pyplot as plt
import os

import pandas as pd
```

```python
data = pd.read_csv("KUplifts.csv")
```

```python
MR = np.array(data.loc[:9,"MR"])
MK = np.array(data.loc[:9,"MK"])
MKe = np.array(data.loc[:9,"MKe"])
LR = np.array(data.loc[:20,"LR"])
LK = np.array(data.loc[:20,"LK"])
LKe = np.array(data.loc[:20,"LKe"])
MeR = np.array(data.loc[:,"MeR"])
MeK = np.array(data.loc[:,"MeK"])
MeKe = np.array(data.loc[:,"MeKe"])
UR = data.loc[0,"UR"]
UK = data.loc[0,"UK"]
UKe = data.loc[0,"UKe"]
SR = data.loc[0,"SR"]
SK = data.loc[0,"SK"]
SKe = data.loc[0,"SKe"]
Ka = data.loc[0,"Ka"]
```

```python
plt.figure(figsize=(6,4),dpi=200);
plt.errorbar(SR, SK, yerr=SKe, fmt="o", label = "Ryugu", capsize=3, alpha = 0.7, marker = "x")
plt.errorbar(LR, LK, yerr=LKe, fmt="o", label = "Moon", capsize=3, alpha = 0.7, marker = "^")
plt.errorbar(MeR*1000, MeK, yerr=MeKe, fmt="o", label = "Mercury", capsize=3, alpha = 0.7, marker = "<")
plt.errorbar(MR, MK, yerr=MKe, fmt="o", label = "Mars", capsize=3, alpha = 0.7, marker = "o")
plt.errorbar(UR, UK, yerr=UKe, fmt="o", label = "Sand, Earth", capsize=3, alpha = 0.7, marker = "v")
plt.errorbar(1200, 50/1200, yerr=40/1200, fmt="o", label = "Meteor Crater, Earth", capsize=3, alpha = 0.7, marker = ">")
#plt.errorbar([12.5, 54, 43], [1.5/12.5, 0.9/54, 0.9/43], yerr = [0.008057394, 0.001852109, 0.002326091], fmt="o", label = "Explosion Craters, Earth", capsize=3, alpha = 0.7, marker = "d")
plt.errorbar([45],[0.05],label = "DART Estimate", alpha = 0.7, marker = "*")
#plt.plot([.01,100000],[Ka,Ka])
plt.xscale("log")
plt.ylabel("$\mathcal{K}_z$")
plt.xlabel("$R_c$ (m)")
plt.legend(prop={'size': 8})
plt.savefig("gravity_invariant.png", dpi=500, )
#ADD DART DATA 
#Note that mercury & SCI data was reduced by 20% to get rid of ejecta blanket thickness
```
